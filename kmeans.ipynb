# Read Dataset Iris

```python
from sklearn import datasets
dataset_iris = datasets.load_iris()
data_iris = dataset_iris.data
print(data_iris)
```

```python
target_iris = dataset_iris.target
target_iris
```

# K-Means

## Clustering

```python
import numpy as np
import kmeans

kmeans = kmeans.KMeans(data_iris, 3, 0)
labels = kmeans.fit()
print(labels)
```

```python
print(kmeans.centroid)
```

## Performance Measure

```python
from sklearn import metrics
from sklearn.metrics.cluster import contingency_matrix

def purity_score(y_true, y_pred):
    cm = contingency_matrix(y_true, y_pred)
    return np.sum(np.amax(cm, axis = 0) / np.sum(cm))

print("Homogeneity: %0.4f" % metrics.homogeneity_score(target_iris, labels))
print("Completeness: %0.4f" % metrics.completeness_score(target_iris, labels))
print("V-measure: %0.4f" % metrics.v_measure_score(target_iris, labels))
print("Purity: %0.4f" % purity_score(target_iris, labels))
```
